```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Transaction ID']
  ys = series['Age']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = _df_15.sort_values('Transaction ID', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Product Category')):
  _plot_series(series, series_name, i)
  fig.legend(title='Product Category', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Transaction ID')
_ = plt.ylabel('Age')
```

```python
# Distribution Plots with Mean and Median
for col in df.select_dtypes('int').columns:
    plt.figure(figsize=(10, 5))
    sns.distplot(df[col], color='blue')

    plt.axvline(df[col].mean(), color='magenta', linestyle='dashed', linewidth = 2)
    plt.axvline(df[col].median(), color='cyan', linestyle='dashed', linewidth = 2)
    plt.xlabel(col, fontsize=13)
    plt.ylabel('count')
    plt.grid()
    plt.show()
```

```python
# Quantity Sold by Product Category
plt.figure(figsize=(10, 6))
df.groupby('Product_Category')['Quantity'].sum().plot(kind='bar')
plt.xticks(rotation=45)
plt.title('Quantity Sold by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
# Sales Distribution in Each Product Category by Gender
plt.figure(figsize=(10, 6))
df.groupby(['Gender', 'Product_Category'])['Customer_ID'].count().unstack().plot(kind='bar', stacked=True)
plt.title('Sales Distribution in Each Product Category by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Product Category')
plt.show()
```

```python
# Age Distribution of Store Customers
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution of Store Customers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Product Category Distribution
plt.figure(figsize=(8, 8))
df['Product_Category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Product Category Distribution')
plt.ylabel(None)
plt.show()
```

```python
# Average Price per Unit by Product Category
plt.figure(figsize=(10, 6))
df.groupby('Product_Category')['Price_per_Unit'].mean().plot(kind='bar', color='purple')
plt.title('Average Price per Unit by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Num_Month'] = df['Date'].dt.month
```

```python
transaction_count = df.groupby('Num_Month')['Transaction_ID'].count()
transaction_count
```

```python
# Monthly Sales Trends
df['Month'] = df['Date'].dt.month_name()
monthly_sales = df.groupby('Month')['Total_Amount'].sum()
plt.figure(figsize=(10, 6))
monthly_sales.plot(marker='o', color='b', linestyle='-')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Transaction Frequency Over Months
plt.figure(figsize=(10, 6))
df['Date'].dt.month.value_counts().sort_index().plot(kind='bar')
plt.title('Transaction Frequency Over Months')
plt.xlabel('Month')
plt.ylabel('Transaction Count')
plt.grid(True)
plt.xticks(rotation=0)
plt.show()
```

```python
# Create a pivot table to analyze sales per product category across months
sales_per_cat = df.pivot_table(index='Num_Month', columns='Product_Category', values='Total_Amount')

# Plotting the sales per product category across months
plt.figure(figsize=(12, 8))
sales_per_cat.plot(kind='bar', figsize=(12, 8))

# Set plot title and labels
plt.title('Monthly Sales per Product Category', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Sales Amount', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=0)

# Place legend in upper left corner
plt.legend(loc='upper left')

# Display the plot
plt.show()
```

```python
def create_age_group(age):
    if age <= 18:
        return '0-18'
    elif age <= 25:
        return '19-25'
    elif age <= 35:
        return '26-35'
    elif age <= 50:
        return '36-50'
    else:
        return '50+'

df['Age_Group'] = df['Age'].apply(create_age_group)
df.head()
```

```python
# Product Category purchased by Age group
plt.figure(figsize=(10, 6))
sns.barplot(x='Age_Group', y='Quantity', hue='Product_Category', data=df)
plt.title('Product Category purchased by Age group')
plt.xlabel('Age Group')
plt.ylabel('Quantity Purchased')
plt.show()
```

```python
# Correlation Between Age and Total Spending
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Total_Amount'], alpha=0.5)
plt.title('Correlation Between Age and Total Spending')
plt.xlabel('Age')
plt.ylabel('Total Amount')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Grouping data by month and calculating total sales amount
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Total_Amount'].sum()

# Plotting the time series data
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index.to_timestamp(), monthly_sales.values, marker='o', color='b', linestyle='-')
plt.title('Monthly Total Sales Amount')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
